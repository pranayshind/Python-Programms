```python
class Player:
    def __init__(self,ID,S,L,P):
        self.ID=ID
        self.S=S
        self.L=L
        self.P=P
class Solution:
    @staticmethod
    def findPointsForGivenSkill(l,S1):
        l1=[]
        for i in l:
            if i.S.lower()==S1.lower():
                l1.append(i.P)
        if l1==[]:
            return None
        else:
            return sum(l1)
    @staticmethod
    def getPlayerBasedOnLevel(l,S1,L1):
        l2=[]
        for i in l:
            if i.S.lower()==S1.lower() and i.L.lower()==L1.lower() and i.P>=20:
                l2.append(i.ID)
        if l2==[]:
            return None
        else:
            return l2
        
n=int(input())
l=[]
for i in range(n):
    ID=int(input())
    S=input()
    L=input()
    P=int(input())
    l.append(Player(ID,S,L,P))
S1=input()
L1=input()
o1=Solution()
o2=o1.findPointsForGivenSkill(l,S1)
if o2==None:
    print("The given Skill is not available")
else:
    print(o2)
o3=o1.getPlayerBasedOnLevel(l,S1,L1)
if o3==None:
    print("No player is available with specified level, skill and eligibility points")
else:
    for i in o3:
        print(i)
    
```

```python
class Movie:
    def __init__(self,ID,D,R,B):
        self.ID=ID
        self.D=D
        self.R=R
        self.B=B
class Solution:
    @staticmethod
    def findAvgBudgetByDirector(l,D):
        l1=[]
        for i in l:
            if i.D.lower()==D.lower():
                l1.append(i.B)
        if l1==[]:
            return None
        else:
            return sum(l1)/len(l1)
        
    @staticmethod
    def getMovieByRatingBudget(l,R1,B1):
        l2=[]
        for i in l:
            if i.R==R1 and i.B==B1:
                l2.append(i.ID)
        if l2==[]:
            return None:
        else:
            return l2
n=int(input())
l=[]
for i in range(n):
    ID=int(input())
    D=input()
    R=int(input())
    B=int(input())
    l.append(Movie(ID,D,R,B))
o1=Solution()
o2=o1.findAvgBudgetByDirector(l,D)
if o2==None:
    print("Sorry - The given director has not yet directed any movie")
else:
    print(o2)
o3=o1.getMovieByRatingBudget(l,R1,B1)
if o2==None:
    print("Sorry - No movie is available with the specified rating and budget requirement")
else:
    for i in o3:
        print(i)
```
